```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from keras.layers import *
from keras.models import *
from keras.utils import *
from sklearn.preprocessing import *
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import random
import re
```

```python
file = './data/text.txt'
with open(file) as f:
    lines = f.readlines()
    for i in lines:
        print(i)
```

```python

text=''
with open(file) as f:
    lines = f.readlines()
    text = text.join([l for l in lines if re.match(r'^[A-Z].*:', l)])

text
    
```

```python
array = []
for t in text.split('\n')[:1000]:
    array.append(t)
```

```python
token = Tokenizer(lower=False, filters='.,?;"-_')
token.fit_on_texts(array)
```

```python
token.word_index
```

```python
len(token.word_index)
```

```python
sequence = token.texts_to_sequences(array)
```

```python
sequence
```

```python
sequence = pad_sequences(sequence, maxlen=10)
sequence


```

```python
X = sequence
Y = np.vstack((X[1:], X[0]))
Y
```

```python
Y.shape
```

```python
# reshape
X = X.reshape(-1, 10, 1)
Y = Y.reshape(-1, 10, 1)
X
```

```python
Y = to_categorical(Y)
Y
```

```python
Y.shape
```

```python
# modeling
model = Sequential()
model.add(Bidirectional(SimpleRNN(128, return_sequences=True), input_shape=(10,1)))
model.add(Bidirectional(SimpleRNN(128, return_sequences=True)))
model.add(Dense(1440, activation='softmax'))
```

```python
model.summary()
```

```python
# compile
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# training
model.fit(X, Y, epochs=100, batch_size=1)
```

```python
# model save
model.save('./data/temp_model')
```
